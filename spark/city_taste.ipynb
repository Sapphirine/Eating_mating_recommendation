```python
# Load and parse the data
data = sc.textFile("data/yelp_dataset_challenge_academic_dataset/yelp_academic_dataset_business.json")
```

```python
business = data.map(lambda r: json.loads(r)).map(lambda r: (r['business_id'], {
            "bid": r['business_id'],
            "city": r['city'],
            "rating": r['stars'],
            "reviews": r['review_count'],
            "categories": r['categories']
        })).cache()
```

```python
city = set(["Charlote", "Charlotte", "Charlotte "])
```

```python
city_data = business.filter(lambda r: r[1]['city'] in city and r[1]['rating'] > 3.5 and 'Restaurants' in r[1]['categories']).cache()
```
